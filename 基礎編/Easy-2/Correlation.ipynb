$$
相関係数\quad r = \frac{ \frac{1}{n}  \sum_{i=1}^{n} (x_i - m_x)(y_i - m_y)}
{ \sqrt {\frac{1}{n}  \sum_{i=1}^{n} (x_i - m_x)^{ 2 }} \times \sqrt {\frac{1}{n}  \sum_{i=1}^{n} (y_i - m_y)^{ 2 }}}
$$
