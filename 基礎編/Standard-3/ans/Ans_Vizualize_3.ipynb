### 0. 必要なモジュールをimport してください（必要に応じて以下のセルに追記していくこと）

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib
%matplotlib inline
```

### 1. dataディレクトリにある、eStat_子供の学習費調査_平成３０年度_学校種別の学習費_加工済み.csvを読み込み、データフレームとして出力してください。

```python
df = pd.read_csv('./data/eStat_子供の学習費調査_平成３０年度_学校種別の学習費_加工済み.csv')
```

```python
df
```

### 3. 学習費総額, 学校教育費, 授業料, 修学旅行・遠足・見学費, 学級・児童会・生徒会費, PTA会費の行を抜き出し、新しいデータフレームとしてください。

```python
tmp = df.iloc[:6, :]
```

### 4. 表頭、表側を入れ替えたデータフレームを出力してください。
なお、表側(index)は学校種別にし、表頭（columns）は学習費総額, 学校教育費, 授業料, 修学旅行・遠足・見学費, 学級・児童会・生徒会費, PTA会費としてください。

```python
tmp = tmp.T.reset_index()
tmp.drop(index=0, inplace=True, axis=0)
tmp.columns = ['種別', '学習費総額', '学校教育費', '授業料', '修学旅行・遠足・見学費', '学級・児童会・生徒会費', 'PTA会費']
```

```python
tmp
```

```python
tmp.set_index('種別')
```

### 5. 学校種別毎の授業料を図示してください。

```python
fig, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x='種別', y='授業料', data=tmp, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
```

### 6. 学校種別毎の学校教育費を図示してください

```python
fig, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x='種別', y='学校教育費', data=tmp, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
```

### 7. 学校種別毎のPTA会費を図示してください

```python
fig, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x='種別', y='PTA会費', data=tmp, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
```
